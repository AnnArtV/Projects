```python
import numpy as np 
import pandas as pd 
import sqlite3
import matplotlib.pyplot as plt
```

```python
database = "../Projects/database.sqlite" # присвоение переменной значение, равное пути до файла базы данных
```

```python
# запрос для получения списка всех таблиц в базе данных SQLite
connect = sqlite3.connect(database) # подключение к базе данных "database.sqlite", используя модуль sqlite3 

Tables = pd.read_sql("""SELECT *     
                        FROM sqlite_master
                        WHERE type='table';""", connect)
                                                      
Tables # результат запроса = таблица, содержащая названия всех доступных таблиц из базы данных, представленные в виде строк и столбцов
```

```python
# извлечение списка стран из таблицы "Country"

Countries = pd.read_sql("""SELECT *
                        FROM Country;""", connect)
Countries
```

```python
# соединение таблицы "League" и "Country" по полю "id" таблицы "Country" и "country_id" таблицы "League"
# результатом запроса является список лиг и их стран
Leagues = pd.read_sql("""SELECT *
                        FROM League
                        JOIN Country ON Country.id = League.country_id;""", connect) 
Leagues
```

```python
# список команд по возрастанию
# ORDER BY определяет сортировку выходных данных - по возрастанию или по убыванию (DESC)
# LIMIT ограничивает количество строк в выходных данных - после сортировки
Teams = pd.read_sql("""SELECT *
                        FROM Team
                        ORDER BY team_long_name
                        LIMIT 10;""", connect)
Teams
```

```python
# список матчей в Испании
# база данных спроектирована в звездообразной структуре - одна таблица (Match) со всеми "показателями" и метриками в форме ключей и идентификаторов,
# в то время как вся описательная информация хранится в других таблицах (Country, league, Team)
# выполняется SQL-запрос, который выбирает данные из таблицы Match и объединяет с таблицами Country, League, Team
# в запросе используются соединения JOIN и LEFT JOIN для объединения таблиц и установления соответствий по заданным условиям
# последние 10 записей, удовлетворяющих условию "country_name = 'Spain'", упорядоченные по дате, сохраняются в датафрейме List_of_matches

List_of_matches = pd.read_sql("""SELECT Match.id, 
                                        Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season, 
                                        stage, 
                                        date,
                                        HT.team_long_name AS  home_team,
                                        AT.team_long_name AS away_team,
                                        home_team_goal, 
                                        away_team_goal                                        
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name = 'Spain'
                                ORDER by date
                                LIMIT 10;""", connect)
List_of_matches
```

```python
# фильтр WHERE, который выбирает только строки со следующими странами:
# Spain', 'Germany', 'France', 'Italy', 'England'
# Группируются данные по Country.name, League.name, season
# Применяется условие HAVING, которое выбирает только группы, где количество distinct stage больше 10
# Результаты сортируются по Country.name, League.name, season в порядке убывания
leages_by_season = pd.read_sql("""SELECT Country.name AS country_name, 
                                        League.name AS league_name, 
                                        season,
                                        count(distinct stage) AS number_of_stages,
                                        count(distinct HT.team_long_name) AS number_of_teams,
                                        avg(home_team_goal) AS avg_home_team_scors, 
                                        avg(away_team_goal) AS avg_away_team_goals, 
                                        avg(home_team_goal-away_team_goal) AS avg_goal_dif, 
                                        avg(home_team_goal+away_team_goal) AS avg_goals, 
                                        sum(home_team_goal+away_team_goal) AS total_goals                                       
                                FROM Match
                                JOIN Country on Country.id = Match.country_id
                                JOIN League on League.id = Match.league_id
                                LEFT JOIN Team AS HT on HT.team_api_id = Match.home_team_api_id
                                LEFT JOIN Team AS AT on AT.team_api_id = Match.away_team_api_id
                                WHERE country_name in ('Spain', 'Germany', 'France', 'Italy', 'England')
                                GROUP BY Country.name, League.name, season
                                HAVING count(distinct stage) > 10
                                ORDER BY Country.name, League.name, season DESC
                                ;""", connection)
leages_by_season
```

```python
# построение графика "Average Goals per Game Over Time"
df = pd.DataFrame(index=np.sort(leages_by_season['season'].unique()), columns=leages_by_season['country_name'].unique())

df.loc[:,'Germany'] = list(leages_by_season.loc[leages_by_season['country_name']=='Germany','avg_goals'])
df.loc[:,'Spain']   = list(leages_by_season.loc[leages_by_season['country_name']=='Spain','avg_goals'])
df.loc[:,'France']   = list(leages_by_season.loc[leages_by_season['country_name']=='France','avg_goals'])
df.loc[:,'Italy']   = list(leages_by_season.loc[leages_by_season['country_name']=='Italy','avg_goals'])
df.loc[:,'England']   = list(leages_by_season.loc[leages_by_season['country_name']=='England','avg_goals'])

df.plot(figsize=(12,5),title='Average Goals per Game Over Time')
```

```python
# извлечение данных о росте игроков из таблицы PLAYER
# в случае если рост игрока округленный ниже 165, устанавливает значение роста равным 165
# в случае если рост игрока округленный выше 195, устанавливает значение роста равным 195
# Если рост игрока находится в промежутке между 165 и 195, оставляет значение роста без изменений
# Подсчитывает количество игроков с каждым из ростовых интервалов и записывает его в столбец 'distribution'.
# Вычисляет среднее значение overall_rating и potential для каждого ростового интервала и записывает их в столбцы 'avg_overall_rating' и 'avg_potential'
# Вычисляет среднее значение веса игроков и записывает его в столбец 'avg_weight'
# Группирует результаты по ростовым интервалам
# Сортирует результаты по возрастанию роста.
players_height = pd.read_sql("""SELECT CASE
                                        WHEN ROUND(height)<165 then 165
                                        WHEN ROUND(height)>195 then 195
                                        ELSE ROUND(height)
                                        END AS calc_height, 
                                        COUNT(height) AS distribution, 
                                        (avg(PA_Grouped.avg_overall_rating)) AS avg_overall_rating,
                                        (avg(PA_Grouped.avg_potential)) AS avg_potential,
                                        AVG(weight) AS avg_weight 
                            FROM PLAYER
                            LEFT JOIN (SELECT Player_Attributes.player_api_id, 
                                        avg(Player_Attributes.overall_rating) AS avg_overall_rating,
                                        avg(Player_Attributes.potential) AS avg_potential  
                                        FROM Player_Attributes
                                        GROUP BY Player_Attributes.player_api_id) 
                                        AS PA_Grouped ON PLAYER.player_api_id = PA_Grouped.player_api_id
                            GROUP BY calc_height
                            ORDER BY calc_height
                                ;""", conn)
players_height
```
